```python
import pandas as pd
import numpy as np

train = pd.read_csv("./pre_processed_train.csv",index_col="id")
test = pd.read_csv("./pre_processed_test.csv",index_col="id")

X=train.copy()
X_test=test.copy()


y=X.label
y_test=X_test.label

X=X.drop(['label'],axis=1)
X_test=X_test.drop(['label'],axis=1)

train.head()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

import warnings
warnings.filterwarnings(action='ignore', category=UserWarning)
```

```python
# evaluate a given model using cross-validation
def evaluate_model(model, X, y):
	cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=4, random_state=1)
	scores = cross_val_score(model, X, y, scoring='f1', cv=cv, n_jobs=-1, error_score='raise')
	return scores
```

```python
from lightgbm import LGBMClassifier


model = LGBMClassifier()
scores = evaluate_model(model, X, y)

from numpy import mean
from numpy import std
print('>%s %.2f (%.2f)' % ('LGBM', scores.mean(), std(scores)))

```

```python
def get_models():
    
    models = dict()
    models['lr'] = LogisticRegression()
    models['svm'] = SVC()
    models['xgboost'] = XGBClassifier()
    models['GBM'] = GradientBoostingClassifier()
    models['LGBM'] = LGBMClassifier()
    
    
    return models
```
